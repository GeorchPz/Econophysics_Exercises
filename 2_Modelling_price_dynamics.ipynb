# Generate random numbers distributed according to a particular distribution

## Uniform distribution

The probability density function of the **uniform** distribution is given by

\begin{equation}
f(x)= \begin{cases}\frac{1}{b-a} & \text { for } a \leq x \leq b \\ 0 & \text { for } x<a \text { or } x>b\end{cases}
\end{equation}

To generate uniformly distributed random numbers we can use the **numpy.random.uniform(a,b, N)** function


```python
a = 0
b = 1

N = 10**6

u = np.random.uniform(a, b, N)

plt.hist(u, bins=100, density=True);
```

## Gaussian distribution

The probability density function of the **Gaussian** distribution is given by

\begin{equation}
f(x)=\frac{1}{\sigma \sqrt{2 \pi}} \exp\left[-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2\right]
\end{equation}

where $\mu$ is the mean and $\sigma$ is the standard deviation.

To generate uniformly distributed random numbers we can use the **numpy.random.normal(μ,σ, N)** function


```python
def gaussian(x, mu, sigma):

  return 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/(2*sigma**2))

mu = 0
sigma = 1

N = 5*10**6

n = np.random.normal(mu, sigma, N)

x = np.linspace(np.amin(n), np.amax(n), 1000)

plt.hist(n, bins=100, density=True)
plt.plot(x, gaussian(x, mu, sigma), lw=3, color="r")
```

## Exponential distribution

The probability density function of the **exponential** distribution is given by

\begin{equation}
f(x)=\exp (-x)
\end{equation}

but it is usually written as

\begin{equation}
f(x; \lambda)=\lambda \exp (-\lambda x)
\end{equation}

Both formulations are related by

\begin{equation}
f(x; \lambda)=f\lambda(\lambda x)=\lambda\exp (-\lambda x)
\end{equation}

To generate exponentially distributed random numbers we can use the **scipy.stats.expon(size=N)** function. Furthermore, the pdf can be easily plotted using **stats.expon.pdf(x)** function.


```python
N = 5*10**6

e = stats.expon.rvs(size=N)

x = np.linspace(np.amin(e), np.amax(e), 1000)

plt.hist(e, bins=100, density=True)
plt.plot(x, stats.expon.pdf(x), lw=3, color="r")
```

## Beta distribution

The probability density function of the **Beta** distribution is given by

\begin{equation}
f(x; a, b)=\frac{\Gamma(a+b) x^{a-1}(1-x)^{b-1}}{\Gamma(a) \Gamma(b)} \,
\end{equation}

where $a$ and $b$ are shape parameters and $\Gamma$ is the gamma function.

To generate exponentially distributed random numbers we can use the **scipy.stats.beta(a, b, size=N)** function. Furthermore, the pdf can be easily plotted using **stats.beta.pdf(x, a, b)** function.


```python
a = 2
b = 5

N = 5*10**6

beta = stats.beta.rvs(a, b, size=N)

x = np.linspace(np.amin(beta), np.amax(beta), 1000)

plt.hist(beta, bins=100, density=True)
plt.plot(x, stats.beta.pdf(x, a, b), lw=3, color="r")
```

## Gamma distribution

The probability density function of the **Gamma** distribution is given by

\begin{equation}
f(x; a)=\frac{x^{a-1} e^{-x}}{\Gamma(a)}
\end{equation},

where $a$ is the shape parameter. It can also be found as

\begin{equation}
f(x ; \alpha, \beta)=\frac{x^{\alpha-1} e^{-\beta x} \beta^\alpha}{\Gamma(\alpha)} \ ,
\end{equation}

where $\alpha$ and $\beta$ are denoted as the shape and rate parameters.

Both formulations are related by

\begin{equation}
f(x ; \alpha, \beta)=\beta \cdot f(\beta x; a)=\beta\frac{\beta^{\alpha-1}x^{\alpha-1}e^{-\beta x}}{\Gamma{a}}=\frac{x^{\alpha-1} e^{-\beta x} \beta^\alpha}{\Gamma(\alpha)}
\end{equation}

```python
a = 2

N = 5*10**6

gamma = stats.gamma.rvs(a, size=N)

x = np.linspace(np.amin(gamma), np.amax(gamma), 1000)

plt.hist(gamma, bins=100, density=True)
plt.plot(x, stats.gamma.pdf(x, a), lw=3, color="r")
```

## Student's t distribution

The probability density function of the Student's t distribution is given by

\begin{equation}
f(x; \nu)=\frac{\Gamma((\nu+1) / 2)}{\sqrt{\pi \nu} \Gamma(\nu / 2)}\left(1+x^2 / \nu\right)^{-(\nu+1) / 2} \ ,
\end{equation}

where $df$ is the number of degrees of freedom.

```python
df = 10

N = 5*10**6

t = stats.t.rvs(df, size=N)

x = np.linspace(np.amin(t), np.amax(t), 1000)

plt.hist(t, bins=100, density=True)
plt.plot(x, stats.t.pdf(x, df), lw=3, color="r")
```

Note that the limit of the Student's t distribution when the degrees f freedom go to infinite is the normal distribution



```python
df_1 = 5
df_2 = 10
df_3 = 50
df_4 = 5000


x = np.linspace(-5, 5, 1000)

plt.figure(figsize=(8, 6))

plt.plot(x, stats.t.pdf(x, df_1), lw=3, color="C0", label="df=%i" % df_1)
plt.plot(x, stats.t.pdf(x, df_2), lw=3, color="C1", label="df=%i" % df_2)
plt.plot(x, stats.t.pdf(x, df_3), lw=3, color="C2", label="df=%i" % df_3)
plt.plot(x, stats.t.pdf(x, df_4), lw=3, color="C3", label="df=%i" % df_4)

plt.plot(x, gaussian(x, 0, 1), color="k", ls="--", lw=3, label="N(0,1)")

plt.yscale("log")

plt.legend();
```

#Fitting a data sample to a distribution

To fit a particular distribution to a given data sample we can use the **scipy.stats.fit** function, which used the Maximum Likelihood Estimation (MLE) method to estimate the parameters of the distribution.

It works both with continuous and discrete distributions, estimating their probability density function (pdf) and probability mass function (pmf), respectively.

```python
#Generate some random numbers according to a Binomial distribution
n = 10
p = 0.3

N = 10**5

r = stats.binom.rvs(n, p, size=N)

#Generate some random numbers according to a beta distribution
a = 1
b = 5

a2 = 5
b2 = 3

N = 10**5

r2 = stats.beta.rvs(a,b, size=N)
r3 = stats.beta.rvs(a2,b2, size=N)

#Fit the samples to the binomial and beta distributions
shape_bounds = {'n': (0, 20), 'p': (0, 1)}
shape_bounds_2 = {'a': (0, 10), 'b':(0, 10)}

fit = stats.fit(stats.binom, r, shape_bounds)
fit2 = stats.fit(stats.beta, r2, shape_bounds_2)
fit3 = stats.fit(stats.beta, r3, shape_bounds_2)

print("---- Fit of Binomial distribution ----\n")
print(fit)

print("\n ---- Fit of beta distribution ----\n")
print(fit2)

print("\n ---- Fit of beta distribution (2) ----\n")
print(fit3)
print()

fig, ax = plt.subplot_mosaic("""AB""", figsize=(8*2, 6))

n_fit, p_fit = fit.params[0], fit.params[1]

x = np.arange(0, 11, 1)

ax["A"].hist(r, bins=np.arange(0, n+2, 1), density=True);
ax["A"].plot(x, stats.binom.pmf(x, n_fit, p_fit), ls="-", marker="o", color="r", ms=10)

ax["A"].set_title("Fit to a Binomial distribution")
ax["A"].set_xlabel("x")
ax["A"].set_ylabel("PDF")

a_fit, b_fit = fit2.params[0], fit2.params[1]
a2_fit, b2_fit = fit3.params[0], fit3.params[1]

x2 = np.linspace(0, 1, 100)

ax["B"].hist(r2, bins=100, density=True);
ax["B"].plot(x2, stats.beta.pdf(x2, a_fit, b_fit), color="r", lw=3)

ax["B"].hist(r3, bins=100, density=True);
ax["B"].plot(x2, stats.beta.pdf(x2, a2_fit, b2_fit), color="r", lw=3)

ax["B"].set_title("Fit to a Beta distribution")

ax["B"].set_xlabel("x")
ax["B"].set_ylabel("PDF")
```

#Exercise 2: Simulation of the 1D Wiener process (Brownian motion)

The Wiener process (or Brownian motion) can be understood as the continuum limit of the Random Walk (RW) process. The stochastic process can be characterized with the following equation

\begin{equation}
   X(t + dt) = X(t) + \mathcal{N}(0, D\sqrt{dt})
\end{equation}

1. Create a function that implements the brownian motion

  **Clue:** *All the random numbers can be generated at once using numpy or scipy. The numpy.cumsum(x, axis) method returns the cumulative sum of the array x over the given axis*.



Since Brownian motion can be seen as the limit of the Random Walk (with an infinite number of infinitesimally small steps), the normal distribution is the go to choice because of the Central Limit Theorem, with a mean $\mu = 0$ and a variance $\sigma^2 = 2 D t $.

```python
def brownian_motion(T, dt, D):
  num_steps = int(T / dt)
  mu = 0 # Mean (always 0, there is no drift)
  sigma = np.sqrt(2 * D * dt)

  # Generate random steps with a Gaussian distribution
  steps = np.random.normal(mu, sigma, num_steps)
  # Brownian motion path
  bm = np.cumsum(steps)
  # Include the starting point (0 in position 0)
  bm = np.insert(bm, 0, 0)

  return bm

brownian_motion(T= 10, dt= 1e-3, D=1)
```

2. Generate and plot 100 trajectories of the brownian motion together with the mean trajectory using $dt=10^{-3}$, $T=10$ and D=1 .


```python
T= 10; dt= 1e-3; D=1

# Generate 100 trajectories
bm_trajectories = np.array([brownian_motion(T, dt, D) for _ in range(100)])
# Mean trajectory
mean_trajectory = np.mean(bm_trajectories, axis=0)

# Plotting
plt.figure(figsize=(14, 7))
for traj in bm_trajectories:
    plt.plot(np.linspace(0, T, int(T/dt) + 1), traj, linewidth=0.5, alpha=0.3)
plt.plot(np.linspace(0, T, int(T/dt) + 1), mean_trajectory, color='red', linewidth=2, label='Mean Trajectory')
plt.title('100 Trajectories of Brownian Motion')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()
```

3. Compute the root mean squared (RMS) displacement for different values of $T$. How does the RMS displacement relates to the total time $T$?

  **Clue:** *The displacement of one trajectory is defined as $\lVert{\vec{x}-\vec{x}_0}\rVert$

$x_{RMS} = \sqrt{\langle ( x−x_0​ )^2 \rangle}$

Theoretical Trend:
$x_{RMS} (N \to \infty) = \sqrt { 2 D T }$


```python
def rms_displacement(T, dt=1e-3, D=1):
  traj = brownian_motion(T, dt, D)
  displacement = traj[-1] - traj[0]  # x - x0
  rms = np.sqrt(np.mean(displacement**2))
  return rms

# T & RMS Displacement values
T_values = np.arange(1,21)
rms_values = [rms_displacement(T) for T in T_values]

# Mean RMS
def mean_rms_displacement(T, dt=1e-3, D=1, num_trajectories= 200):
  rms_displacements = [rms_displacement(T, dt, D) for _ in range(num_trajectories)]
  return np.mean(rms_displacements)

mean_rms_values = [mean_rms_displacement(T) for T in T_values]


# Plotting
plt.figure(figsize=(16, 8))
# 1 trajectory
plt.plot(T_values, rms_values, marker='o', linestyle='-', label='Single Trajectory')
# Mean of multiple traj
plt.plot(T_values, mean_rms_values, marker='o', linestyle='-', label='Mean')
# Theory
plt.plot(T_values, np.sqrt(2 * T_values), linestyle='--', color='red', label='Theoretical Trend $(\sqrt{2DT})$')

plt.legend()
plt.grid()
plt.xlabel('T')
plt.ylabel('$x_{RMS}$')
plt.title('RMS Displacement vs Total Time')
plt.show()
```

**How does the RMS displacement relates to the total time T?**

Without doing the mean, its difficult to find a trend because of the diffusive nature of the Brownian Motion.

Once the mean of 200 trajectories was done, we can see that the RMS displacement increases with the total time. This trend can be deduced from the last figure: *100 Trajectories of Brownian Motion*.

Theoretically, the RMS displacement for Brownian motion should scale as $\sqrt{T}$. But the mean lags behind, surely because 200 is quite far from $\infty$.

# Exercise 3: A brownian motion model for stock prices

We can assume that the return of stock prices follow a gaussian distribution of given mean $\mu$ and variance $\sigma^2$ and simulate the possible future evolution of the stock price.

1. Download the data from a stock market of interest (e.g. META, TSLA, MSFT, etc) for the period 2020-2021 with daily resolution

2. Compute the mean and variance of the log_returns.

3. Simulate future log returns using a Brownian motion with the estimated mean and variance. Basically the changes in log returns will be normally sistributed with mean $\mu$ and variance $\sigma^2$.

4. Transform the simulated log returns to new prices. In essence, the log returns fulfill the following relation

\begin{equation}
  S(t)=S(0)e^{rt} \ ,
\end{equation}
where r is the log return.


**Clue:** *To transform the simulated returns into new prices efficiently (in Python) you can make use of a property of the log returns (because of the exponentials):*

\begin{equation}
  S(t+2\Delta t)=S(t+\Delta t)e^{r_1\Delta t}=S(t)e^{r_0\Delta t}e^{r_2\Delta t}=S(t)e^{(r_1+r_0)\Delta t} \ ,
\end{equation}

5. Generate several (e.g. 10) simulations of the future prices for a trading year (252 days) and plot it together with the real evolution that prices followed. What do you think about it?

From the clue, we obtain:

$S(t) = S(0) e^{\sum r_i \Delta t}$

we can calculate the price for each time using a cumulative sum.

```python
start = datetime(2020, 1, 1)
end = datetime(2021, 1, 1) #.today()

# num:      0       1       2       3       4        5          6
asset = ["AAPL", "MSFT", "AMZN", "TSLA", "META", "BTC-USD", "ETH-USD"][5] # [num] -> asset selection

# Download stock data
df = yf.download([asset], start=start, end=end, interval= '1d', progress=False)
df['LogRet'] = np.log( 1 + df['Adj Close'].pct_change() )
df_fit = df.dropna()

# Mean & Standard Deviation (Fit to Gaussian)
mu, sigma = stats.norm.fit(df_fit['LogRet'])

print(f"Mean: {mu}")
print(f"Std:  {sigma}")

T = df_fit.shape[0] # Trading days in a year (whole year for BTC)
num_iterations = 10

# Simulate future log returns
future_log_returns = np.array([
    np.random.normal(mu, sigma, T) for _ in range(num_iterations)
    ])
#future_log_returns.shape

# Starting price (last observed price)
S_0 = df_fit['Adj Close'][-1]

cumulative_future_log_returns = np.cumsum(future_log_returns, axis=1) # along the time axis
future_prices = S_0 * np.exp(cumulative_future_log_returns)
#future_prices.shape

# Date range
dates = pd.date_range(start=end, periods=T)
# real BTC price data
df_real = yf.download(asset, start=start, end=datetime(2022, 1, 1), interval='1d', progress=False)


# Plotting
fig, axes = plt.subplots(2, 1, figsize= (17, 14))

def subplot_simulation(ax, scale):
  # Simulated future price paths
  for i in range(num_iterations):
    ax.plot(dates, future_prices[i], alpha=0.5, zorder=1) # Higher zorder elements appear on top
  # Real historical prices
  ax.plot(df_real['Adj Close'], label='Real Price', color='k', linewidth=2, zorder=2)
  # Starting point of the simulations
  ax.scatter(end, S_0, color='red', s=50, label='Start of Simulations', zorder=3)

  ax.set(ylabel="Price", yscale=scale)
  ax.xaxis.grid()
  ax.legend()

subplot_simulation(axes[0], 'linear')
subplot_simulation(axes[1], 'log')
plt.suptitle(f"Simulated Future Prices for {asset}", y=0.92)
plt.show()
```

**What do you think about the simulated prices?**

The simulated future prices for BTC-USD show a range of outcomes, all of these show a bigger price increase than what was observed in reality after a year. I guess its due to the sharp price increase on the period where the simulation was started, therefore the simulations might be overemphasizing past trends.

However, the simulations appear to do a good job on the first quarter of 2021, it appears that they provide better approximations for short-term predictions. Specially since we are using the approximation of constant volatility.

# Exercise 4.  Data-driven price dynamics model using the Student's t distribution

Another possibility for modelling price dynamics is performing a data-driven simulation. A simple approach is to fit the return distribution of underlying data that one wants to model and simulate future returns according to the fitted distribution.

**1. Fit the distribution of returns using a t distribution**

* Plot the pdf of the fitted distribution together with the histogram of the returns.

**Clue:** *Use the `scipy.stats.t.fit` method to fit the distribution to the data. Note that this method returns a tupple of shape, loc and scale parameters for the fitted distribution. You can then use the `empirical_dist = scipy.stats.t(fited_params)` method to define the empirical distribution of returns. and then use `empirical_dist.pdf(x)` to plot the pdf of the fitted distribution.*

```python
# Fit to t distribution
fitted_params = stats.t.fit(df_fit['LogRet'])
# Distribution with fitted parameters
empirical_dist = stats.t(*fitted_params)

# Range of values for the PDF
r = np.linspace(min(df_fit['LogRet']), max(df_fit['LogRet']), 200)


# Plotting
plt.figure(figsize= (14, 7))
# Histogram of the returns
plt.hist(df_fit['LogRet'], bins=50, density=True, alpha=0.5, label='Real Log Returns')
# PDF of the fitted t distribution
plt.plot(r, empirical_dist.pdf(r), 'r-', lw=2, label='t Distribution')

plt.title("Fitted Student's t Distribution vs Historical Log Returns")
plt.legend()
plt.show()
```

**2. Use the fitted distribution to simulate future returns and future price dynamics**

Plot several realisations of the possible future dynamics of prices.

**Clue:** *You can use the `empirical_dist.rvs(N)* method tho generate random numbers according to the fitted distribution.

```python
# We use the Trading days from before (T)
# Starting price (S_0)
num_iterations = 10
# We won't plot the past trend of the real prices
df_real_cut = df_real[df_real.index >= '2020-11-01']

# Simulate 10 log returns paths
simulated_returns = [empirical_dist.rvs(T) for _ in range(num_iterations)]
# Convert simulated log returns to price paths
simulated_price_paths = [S_0 * np.exp(np.cumsum(sim_ret)) for sim_ret in simulated_returns]


# Plotting
fig, axes = plt.subplots(2, 1, figsize= (14, 14))

def subplot_simulation(ax, scale):
  # Simulated future price paths
  for path in simulated_price_paths:
    ax.plot(dates, path, alpha=0.5, zorder=1)
  # Real historical prices
  ax.plot(df_real_cut['Adj Close'], label='Real Price', color='k', linewidth=2, zorder=2)
  # Starting point of the simulations
  ax.scatter(end, S_0, color='red', s=50, label='Start of Simulations', zorder=3)

  ax.set(ylabel="Price", yscale=scale)
  ax.xaxis.grid()
  ax.legend()

subplot_simulation(axes[0], 'linear')
subplot_simulation(axes[1], 'log')
plt.suptitle(f"Simulated Future Prices for {asset}", y=0.92)
plt.show()
```
